```python
import pandas as pd
```

```python
data_file = '/home/jupyter-julia/storage/lessons/Day3PM_GORG-classifier/cog-20.def.csv'
```

```python
df = pd.read_csv(data_file)
```

```python
df['fc_status'] = [len(i) for i in df['Functional_category']]
df
```

```python
from collections import Counter
Counter(df['fc_status'])
```

```python
ndf = pd.DataFrame(columns = df.columns)

for i, l in df.iterrows():
    fc = l['Functional_category']
  
    if len(fc) == 1:
        ndf.loc[len(ndf.index)] = l
    elif len(l['Functional_category']) > 1:
        for char in fc:
            l['Functional_category'] = char
            ndf.loc[len(ndf.index)] = l
    
```
